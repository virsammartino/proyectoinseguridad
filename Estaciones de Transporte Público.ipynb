### El objetivo es obtener las coordenadas de todas las estaciones de subte, de tren, de premetro y de metrobus

```python
# Importamos librerías
import numpy as np
import pandas as pd
```

```python
# Importamos los datasets
estaciones_tren = pd.read_csv('../Data/estaciones-de-ferrocarril.csv')
estaciones_bus = pd.read_csv('../Data/estaciones-de-metrobus.csv')
estaciones_subte = pd.read_csv('../Data/estaciones-de-subte.csv')
estaciones_premetro = pd.read_csv('../Data/estaciones-premetro.csv')
```

```python
# La columna que muestra las coordenadas para 'estaciones_bus' es conflictiva
WKT_serie = estaciones_bus.WKT
WKT_serie
```

```python
# Quiero sacar "point" y los paréntesis

WKT_serie_strip = WKT_serie.str.strip('POINT ()')
WKT_serie_strip
```

```python
# Quiero separarlo en dos columnas y renombrarlas como "long" y "lat"
WKT_serie_split = WKT_serie_strip.str.split(expand=True)
WKT_serie_split.columns = ['long', 'lat']
WKT_serie_split
```

```python
# Lo agregamos al dataframe original
estaciones_bus = estaciones_bus.merge(WKT_serie_split, left_index=True, right_index=True)
estaciones_bus.loc[:,['WKT','long','lat']].sample(3)
```

```python
estaciones_bus
```

```python
estaciones_bus['WKT'] = estaciones_bus.WKT.str.strip('POINT ')
estaciones_bus
```

```python
estaciones_bus.rename(columns = {'WKT':'lat-lon'}, inplace = True)
estaciones_bus.sample(2)
```
