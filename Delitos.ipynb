```python
# Importamos las librerías
import numpy as np
import pandas as pd
import re
```

```python
# Importamos los datasets
delitos2016 = pd.read_csv('../Data/delitos_2016.csv')
delitos2017 = pd.read_csv('../Data/delitos_2017.csv')
delitos2018 = pd.read_csv('../Data/delitos_2018.csv')
delitos2019 = pd.read_csv('../Data/delitos_2019.csv')
```

```python
# Resumen breve de cada df
print('Delitos 2016')
print(delitos2016.info(),'\n\n')
print('Delitos 2017')
print(delitos2017.info(),'\n\n')
print('Delitos 2018')
print(delitos2018.info(),'\n\n')
print('Delitos 2019')
print(delitos2019.info())
```

```python
# Generamos un dataframe con las columnas de cada una de ellas
delitos_columns = pd.DataFrame([delitos2016.columns, delitos2017.columns, delitos2018.columns, 
                                delitos2019.columns], index=['2016','2017','2018','2019'])
delitos_columns
```

```python
# Agrego el dato del año para cada dataframe
delitos2016['Año'] = 2016
delitos2017['Año'] = 2017
delitos2018['Año'] = 2018
delitos2019['Año'] = 2019
```

```python
# Vemos cómo se ven los datasets de cada año
delitos2016.sample(3)
```

```python
delitos2017.sample(3)
```

```python
delitos2018.sample(3)
```

```python
delitos2019.sample(3)
```

```python
# Observamos que los valores de "fecha" para el dataset de 2019 es distinto al resto, por lo que debemos modificarla

# Agregamos el año al principio y para todos los valores
fecha_full = '2019-'+ delitos2019.fecha
#print(fecha_full)

# Eliminamos la última parte
fecha_clean = fecha_full.str.replace('-19','')
print(fecha_clean)
```

```python
# Modifcamos la columna original
delitos2019['fecha'] = fecha_clean
```

```python
# Vemos nuevamente cómo quedo el df
delitos2019.sample(3)
```

```python
delitos_A = pd.concat([delitos2016,delitos2017,delitos2018,delitos2019], ignore_index=True)
delitos_A = delitos_A.drop(columns=['id'])
delitos_A.info()
```

```python
print('Porcentaje de nulos por columna:')
delitos_A.apply(lambda x: round(x.isnull().sum()/delitos_A.shape[0],2), axis=0)
```

```python
delitos_A.head(5)
```

```python
# Nulos en la columna "comuna"
delitos_A.comuna.isnull().value_counts()
```

```python
# Vemos la cantidad de nulls por año
delitos_A[delitos_A.comuna.isnull()].groupby('Año')['comuna'].size()
```

```python
# Vemos si los nulls de 'comuna' se corresponden con los nulls de 'barrio'
print(all(delitos_A[delitos_A.comuna.isnull()].index == delitos_A[delitos_A.barrio.isnull()].index))
print(all(delitos_A[delitos_A.comuna.isnull()].index == delitos_A[delitos_A.lat.isnull()].index))
print(all(delitos_A[delitos_A.comuna.isnull()].index == delitos_A[delitos_A.long.isnull()].index))
```

```python
# Vemos que los indices de las filas nulas de "Barrio", "Comuna", "Lat" y "lon" son las mismas paras todos
# ¿Podemos imputar a partr de otra columna o haciendo análisis descriptivo de los robos?
```

```python
# Opción 1: Probamos directamente borrando esos datos
delitos_B = delitos_A.drop(index=delitos_A[delitos_A.comuna.isnull()].index)

print('Porcentaje de nulos por columna:')
delitos_B.apply(lambda x: round(x.isnull().sum()/delitos_B.shape[0],2), axis=0)
```

```python
# Vemos los nulos de la columna "cantidad_registrada"
print(delitos_B.cantidad_registrada.value_counts(dropna=False),'\n')
print(delitos_B.cantidad_registrada.value_counts(dropna=False, normalize=True).round(4)*100)
```
